```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr
```

```python
# Load the dataset
data = pd.read_csv('biomedical_data.csv')
```

```python
# Line plot for Biomedical Technologies Introduced over Time
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Biomedical Technologies Introduced'], marker='o', label='Technologies Introduced')
plt.title('Trend of Biomedical Technologies Introduced (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Number of Technologies')
plt.grid(True)
plt.legend()
plt.show()

# Line plot for Health Indicators Improved over Time
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Health Indicators Improved'], marker='s', color='g', label='Health Indicators Improved')
plt.title('Trend of Health Indicators Improved (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Number of Indicators')
plt.grid(True)
plt.legend()
plt.show()

# Line plot for Population Covered over Time
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Population Covered (Millions)'], marker='d', color='r', label='Population Covered (Millions)')
plt.title('Population Coverage Trend (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Population Covered (Millions)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 7))

# Plot each metric
plt.plot(data['Year'], data['Biomedical Technologies Introduced'], marker='o', label='Technologies Introduced', color='blue')
plt.plot(data['Year'], data['Health Indicators Improved'], marker='s', label='Health Indicators Improved', color='green')
plt.plot(data['Year'], data['Population Covered (Millions)'], marker='d', label='Population Covered (Millions)', color='red')

# Add titles and labels
plt.title('Trends in Biomedical Metrics (2015-2023)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Metrics', fontsize=12)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10, loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Pearson Correlation
pearson_inv_pop = pearsonr(data['Govt. Investment (Million USD)'] + data['Private Investment (Million USD)'],
                           data['Population Covered (Millions)'])
pearson_tech_health = pearsonr(data['Biomedical Technologies Introduced'], data['Health Indicators Improved'])
pearson_train_innov = pearsonr(data['Training Programs Conducted'], data['Biomedical Technologies Introduced'])

print("Pearson Correlation Results:")
print(f"Investments vs. Population Coverage: r = {pearson_inv_pop[0]:.2f}, p = {pearson_inv_pop[1]:.2e}")
print(f"Technologies Introduced vs. Health Indicators Improved: r = {pearson_tech_health[0]:.2f}, p = {pearson_tech_health[1]:.2e}")
print(f"Training Programs vs. Innovations: r = {pearson_train_innov[0]:.2f}, p = {pearson_train_innov[1]:.2e}")

# Spearman Correlation
spearman_inv_pop = spearmanr(data['Govt. Investment (Million USD)'] + data['Private Investment (Million USD)'],
                             data['Population Covered (Millions)'])
spearman_tech_health = spearmanr(data['Biomedical Technologies Introduced'], data['Health Indicators Improved'])
spearman_train_innov = spearmanr(data['Training Programs Conducted'], data['Biomedical Technologies Introduced'])

print("\nSpearman Correlation Results:")
print(f"Investments vs. Population Coverage: r = {spearman_inv_pop[0]:.2f}, p = {spearman_inv_pop[1]:.2e}")
print(f"Technologies Introduced vs. Health Indicators Improved: r = {spearman_tech_health[0]:.2f}, p = {spearman_tech_health[1]:.2e}")
print(f"Training Programs vs. Innovations: r = {spearman_train_innov[0]:.2f}, p = {spearman_train_innov[1]:.2e}")
```

```python
# Correlation Heatmap
# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

correlation_matrix = numeric_data.corr()


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

```
